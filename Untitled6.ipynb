```python
import os
import matplotlib.pyplot as plt

# Define paths to the dataset directories
data_dir = "/root/.cache/kagglehub/datasets/kaushalsethia/alzheimers-adni/versions/1"
train_dir = os.path.join(data_dir, "Alzheimers-ADNI/train")
test_dir = os.path.join(data_dir, "Alzheimers-ADNI/test")

# Count files in each class
class_counts = {}
for folder in os.listdir(train_dir):
    folder_path = os.path.join(train_dir, folder)
    if os.path.isdir(folder_path):
        class_counts[folder] = len(os.listdir(folder_path))

# Visualize the class distribution
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Files")
plt.xticks(rotation=45)
plt.show()

print("Class Counts:", class_counts)
```

```python
IMG_HEIGHT = 250  # Image height for resizing
IMG_WIDTH = 250  # Image width for resizing
BATCH_SIZE = 32   # Batch size for training and validation
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define augmentation generator
augmentation = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True
)

# Define writable output directory
output_dir = "augmented_data"
os.makedirs(output_dir, exist_ok=True)

# Augment minority classes
for folder, count in class_counts.items():
    if count < 1000:  # Define a threshold for minority classes
        folder_path = os.path.join(train_dir, folder)
        save_dir = os.path.join(output_dir, f"augmented_{folder}")
        os.makedirs(save_dir, exist_ok=True)

        generator = augmentation.flow_from_directory(
            train_dir,
            classes=[folder],
            target_size=(224, 224),
            batch_size=32,
            save_to_dir=save_dir,
            save_format='jpeg',
            class_mode=None
        )

        # Generate enough samples to reach 1000 images
        target_count = 1000 - count
        for i, _ in enumerate(generator):
            if i >= target_count:
                break
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Prepare labels and counts
classes = list(class_counts.keys())
counts = list(class_counts.values())

# Map classes to indices
label_mapping = {class_name: idx for idx, class_name in enumerate(classes)}
labels = [label_mapping[class_name] for class_name in classes]

# Compute class weights
weights = compute_class_weight('balanced', classes=np.arange(len(classes)), y=np.repeat(labels, counts))
class_weights = dict(enumerate(weights))

print("Class Weights:", class_weights)
```

```python
import shutil

# Define the final dataset directory
final_dataset_dir = "final_dataset"
final_train_dir = os.path.join(final_dataset_dir, "train")
final_test_dir = os.path.join(final_dataset_dir, "test")

# Create the final dataset directories
os.makedirs(final_train_dir, exist_ok=True)
os.makedirs(final_test_dir, exist_ok=True)

# Create subdirectories for each class
all_classes = ["Final AD JPEG", "Final CN JPEG", "Final EMCI JPEG", "Final LMCI JPEG", "Final MCI JPEG"]
for class_name in all_classes:
    os.makedirs(os.path.join(final_train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(final_test_dir, class_name), exist_ok=True)
```

```python
from sklearn.model_selection import train_test_split
import random

# Copy AD and CN data directly
for class_name in ["Final AD JPEG", "Final CN JPEG"]:
    # Train data
    source_train_dir = os.path.join(train_dir, class_name)
    dest_train_dir = os.path.join(final_train_dir, class_name)
    for file_name in os.listdir(source_train_dir):
        shutil.copy(os.path.join(source_train_dir, file_name), os.path.join(dest_train_dir, file_name))

    # Test data
    source_test_dir = os.path.join(test_dir, class_name)
    dest_test_dir = os.path.join(final_test_dir, class_name)
    for file_name in os.listdir(source_test_dir):
        shutil.copy(os.path.join(source_test_dir, file_name), os.path.join(dest_test_dir, file_name))

# Split and copy data for augmented classes
for class_name in ["Final EMCI JPEG", "Final LMCI JPEG", "Final MCI JPEG"]:
    source_dir = os.path.join("augmented_data", f"augmented_{class_name}")
    dest_train_dir = os.path.join(final_train_dir, class_name)
    dest_test_dir = os.path.join(final_test_dir, class_name)

    # Get all file names
    all_files = os.listdir(source_dir)
    train_files, test_files = train_test_split(all_files, test_size=0.2, random_state=42)

    # Copy training files
    for file_name in train_files:
        shutil.copy(os.path.join(source_dir, file_name), os.path.join(dest_train_dir, file_name))

    # Copy testing files
    for file_name in test_files:
        shutil.copy(os.path.join(source_dir, file_name), os.path.join(dest_test_dir, file_name))

print("Final dataset preparation completed!")
```

```python
# Count files in each class for train and test
final_class_counts = {"train": {}, "test": {}}

for dataset_type in ["train", "test"]:
    dataset_dir = os.path.join(final_dataset_dir, dataset_type)
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):
            final_class_counts[dataset_type][class_name] = len(os.listdir(class_path))

print("Final Class Counts (Train):", final_class_counts["train"])
print("Final Class Counts (Test):", final_class_counts["test"])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data generators for training and testing
train_datagen = ImageDataGenerator(rescale=1.0/255)
test_datagen = ImageDataGenerator(rescale=1.0/255)

# Load training data
train_generator = train_datagen.flow_from_directory(
    "final_dataset/train",
    target_size=(250, 250),
    batch_size=32,
    class_mode="categorical"
)

# Load testing data
test_generator = test_datagen.flow_from_directory(
    "final_dataset/test",
    target_size=(250, 250),
    batch_size=32,
    class_mode="categorical"
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(250, 250, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # 5 classes in the dataset
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
# Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Plot training history
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
